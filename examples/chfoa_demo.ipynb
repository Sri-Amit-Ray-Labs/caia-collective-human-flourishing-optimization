# CHFOA Demo Notebook

This notebook demonstrates a simple end-to-end flow using the CHFOA scaffold:

- Generate synthetic data
- Predict flourishing deltas (FSM stub)
- Compute CHFOA scores

The framework is inspired by Sri Amit Ray's Compassionate AI Architecture (CAIA) and Human Flourishing Optimization Algorithm (HFOA).


```python
import pandas as pd
from chfoa.data.synthetic_generator import generate_sample
from chfoa.core.fsm import FSMModel, Context, StakeholderProfile, ActionOption
from chfoa.core.scoring import score_action
```

```python
# Generate synthetic samples
samples = generate_sample(20)
df = pd.DataFrame(samples)
df.head()
```

## Run CHFOA FSM and Scoring

We will:
- Create a simple context and stakeholder profile for each row.
- Call the FSM stub to predict flourishing deltas.
- Compute CHFOA scores using `score_action`.


```python
fsm = FSMModel()

results = []
for i, row in df.iterrows():
    ctx = Context(id=str(row["context_id"]), domain=row["domain"], metadata={})
    stakeholder = StakeholderProfile(
        id=row["stakeholder_group"],
        vulnerability_weight=row["vulnerability_weight"],
        preferences={}
    )
    action = ActionOption(
        id=row["action_type"],
        metadata={},
        features={"task_utility": row["task_utility"]}
    )

    # Predict flourishing (stub)
    pred = fsm.predict(ctx, action, [stakeholder])

    # Use task_utility as base score for demo
    metrics = score_action(base_score=row["task_utility"], pred=pred, stakeholders=[stakeholder])
    results.append({
        "context_id": row["context_id"],
        "stakeholder_group": row["stakeholder_group"],
        "action_type": row["action_type"],
        "base_utility": row["task_utility"],
        "chfoa_score": metrics["score"],
        "flourishing_scalar": metrics["flourishing_scalar"],
        "compassion": metrics["compassion"],
        "harm": metrics["harm"],
        "violations": metrics["violations"]
    })

results_df = pd.DataFrame(results)
results_df.head()
```

## Inspect Scores

Below we compare base task utility vs CHFOA scores. In a full implementation, the FSM and CEE modules will be trained models, and the CHFOA score will reflect real flourishing and compassion effects aligned with Sri Amit Ray CAIA and HFOA.
